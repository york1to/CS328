```python
import subprocess
import matplotlib.pyplot as plt

def run_mpi(num_processes):
    cmd = f"mpirun -np {num_processes} ./mpi_matrix_solved"
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    output = result.stdout

    for line in output.split('\n'):
        if "ELAPSED_TIME:" in line:
            print(f"DEBUG: {line}")
            parts = line.split()
            if len(parts) > 1:
                try:
                    time_taken = float(parts[1])
                    return time_taken
                except ValueError:
                    print(f"Error: Could not convert '{parts[1]}' to float.")
                    return None
            else:
                print("Error: Unexpected format for ELAPSED_TIME line.")
                return None
    return None

process_counts = [2, 4, 8, 16, 32]  # Adjust as needed
times = [run_mpi(p) for p in process_counts]

plt.plot(process_counts, times, marker='o')
plt.title("MPI Computation Time vs Number of Processes")
plt.xlabel("Number of Processes")
plt.ylabel("Computation Time (seconds)")
plt.grid(True)
plt.show()
```
